```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data = {
    'Difficulty': ['easy', 'medium', 'difficult','impossible'],
    'marks': [78, 56, 11,1]
}
df = pd.DataFrame(data)
encoder = LabelEncoder()
df['difficulty_encoder'] = encoder.fit_transform(df['Difficulty'])
print(df)
cor = df['difficulty_encoder'].corr(df['marks'])
print("Correlation:", cor)
```

```python
import pandas as pd
 
# Sample data
data = {'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red'],
        'Test Score': [50, 60, 70, 80, 90]}
 
# Create a DataFrame
df = pd.DataFrame(data)
 
# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Color'])
 
# Show the encoded data
print(df_encoded)
 
# Calculate the correlation
correlation = df_encoded['Color_Red'].corr(df_encoded['Test Score'])
print("Correlation with Red:", correlation)
```
